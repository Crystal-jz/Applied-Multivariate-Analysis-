## (a) 95% T^2 simultaneous confidence intervals for the four population means for length.


```python
n = 7
p = 4

length2 = [141., 140., 145., 146., 150., 142., 139.]
length3 = [157., 168., 162., 159., 158., 140., 171.]
length4 = [168., 174., 172., 176., 168., 178., 176.]
length5 = [183., 170., 177., 171., 175., 189., 175.]

length = np.array([length2, \
                   length3, \
                   length4, \
                   length5, \
                  ]).T

length2_mean = np.mean(length[:, 0])
length3_mean = np.mean(length[:, 1])
length4_mean = np.mean(length[:, 2])
length5_mean = np.mean(length[:, 3])

length_cov = np.cov(length.T)
length_cov

# (a) 95% T^2 simultaneous confidence intervals for the four population means for length.
alpha = 0.05

## 95% T^2 simultaneous confidence intervals for length2
length2_confidence_interval_lower_bound = length2_mean - np.sqrt(length_cov[0,0]/n) * np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))
length2_confidence_interval_upper_bound = length2_mean + np.sqrt(length_cov[0,0]/n) * np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))

## 95% T^2 simultaneous confidence intervals for length3
length3_confidence_interval_lower_bound = length3_mean - np.sqrt(length_cov[1,1]/n) * np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))
length3_confidence_interval_upper_bound = length3_mean + np.sqrt(length_cov[1,1]/n) * np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))

## 95% T^2 simultaneous confidence intervals for length4
length4_confidence_interval_lower_bound = length4_mean - np.sqrt(length_cov[2,2]/n) * np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))
length4_confidence_interval_upper_bound = length4_mean + np.sqrt(length_cov[2,2]/n) * np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))

## 95% T^2 simultaneous confidence intervals for length5
length5_confidence_interval_lower_bound = length5_mean - np.sqrt(length_cov[3,3]/n) * np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))
length5_confidence_interval_upper_bound = length5_mean + np.sqrt(length_cov[3,3]/n) * np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))

print(length2_confidence_interval_lower_bound, length2_confidence_interval_upper_bound)
print(length3_confidence_interval_lower_bound, length3_confidence_interval_upper_bound)
print(length4_confidence_interval_lower_bound, length4_confidence_interval_upper_bound)
print(length5_confidence_interval_lower_bound, length5_confidence_interval_upper_bound)
```

## (b) 95% T^2 simultaneous confidence intervals for the three successive yearly increases in mean length.


```python
# (b) 95% T^2 simultaneous confidence intervals for the three successive yearly increases in mean length.
alpha = 0.05

length_increase = np.array([length[:, 1] - length[:, 0], \
                            length[:, 2] - length[:, 1], \
                            length[:, 3] - length[:, 2], \
                           ]).T

length_increase_mean1 = np.mean(length_increase[:, 0])
length_increase_mean2 = np.mean(length_increase[:, 1])
length_increase_mean3 = np.mean(length_increase[:, 2])


length_increase_cov = np.cov(length_increase.T)
length_increase_cov

## 95% T^2 simultaneous confidence intervals for the increase from 2 to 3 years
length_increase1_confidence_interval_lower_bound = length_increase_mean1 - np.sqrt(length_increase_cov[0,0]/n) *np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))
length_increase1_confidence_interval_upper_bound = length_increase_mean1 + np.sqrt(length_increase_cov[0,0]/n) *np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))

## 95% T^2 simultaneous confidence intervals for the increase from 3 to 4 years
length_increase2_confidence_interval_lower_bound = length_increase_mean2 - np.sqrt(length_increase_cov[1,1]/n) *np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))
length_increase2_confidence_interval_upper_bound = length_increase_mean2 + np.sqrt(length_increase_cov[1,1]/n) *np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))

## 95% T^2 simultaneous confidence intervals for the increase from 4 to 5 years
length_increase3_confidence_interval_lower_bound = length_increase_mean3 - np.sqrt(length_increase_cov[2,2]/n) *np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))
length_increase3_confidence_interval_upper_bound = length_increase_mean3 + np.sqrt(length_increase_cov[2,2]/n) *np.sqrt(((n-1)*p/(n-p))* f.ppf(1-alpha, p, n-p))

print(length_increase1_confidence_interval_lower_bound, length_increase1_confidence_interval_upper_bound)
print(length_increase2_confidence_interval_lower_bound, length_increase2_confidence_interval_upper_bound)
print(length_increase3_confidence_interval_lower_bound, length_increase3_confidence_interval_upper_bound)
```

## (c) 95% T^2 confidence ellipse for the mean increase in length from 2 to 3 years and the mean increase in length from 4 to 5 years.


```python
# (c) 95% T^2 confidence ellipse for the mean increase in length from 2 to 3 years and the mean increase in length from 4 to 5 years.

length_increase_23_and_45 = np.array([length_increase[:, 0], \
                                      length_increase[:, 2], \
                                     ]).T

length_increase_mean1 = np.mean(length_increase_23_and_45[:, 0])
length_increase_mean3 = np.mean(length_increase_23_and_45[:, 1])

length_increase_23_and_45_cov = np.cov(length_increase_23_and_45.T)
length_increase_23_and_45_cov

length_increase__23_and_45_mu = np.array([length_increase_mean1, length_increase_mean3])
length_increase_23_and_45_Sigma = np.array(np.array(length_increase_23_and_45_cov))

length_increase_23_and_45_mu_hat = np.average(length_increase_23_and_45, axis=0)

eigenvalues1, eigenvectors1 = np.linalg.eigh(length_increase_23_and_45_cov)
length_increase_23_and_45_S_square_root = eigenvectors1 @ np.diag(np.sqrt(eigenvalues1)) @ eigenvectors1.T

length_increase_23_and_45_df = pd.DataFrame(length_increase_23_and_45)

length_increase_23_and_45_radius_square = ((n-1)*p/((n-p)*n))* f.ppf(0.95, p, n-p)
length_increase_23_and_45_radius_square

length_increase_ellipsoid = [length_increase_23_and_45_S_square_root @ point + length_increase_23_and_45_mu_hat
                             for point in (np.sqrt(length_increase_23_and_45_radius_square) * circle)
                            ]

ax1 = length_increase_23_and_45_df.plot.scatter(0, 1, c="r", label="$\\vec x_i, i=1,...,n$")
pd.DataFrame(length_increase_ellipsoid).plot.scatter(0, 1, ax=ax1)

pd.DataFrame([(x, -27.) for x in np.linspace(
    length_increase1_confidence_interval_lower_bound,
    length_increase1_confidence_interval_upper_bound
   )]).plot.scatter(0, 1, c='b', ax=ax1, s=100)
pd.DataFrame([(-27., y) for y in np.linspace(
    length_increase3_confidence_interval_lower_bound,
    length_increase3_confidence_interval_upper_bound
   )]).plot.scatter(0, 1, c='b', ax=ax1, s=100)

```

## (a) 90% joint confidence ellipse

```python
n1 = 9
p1 = 2

x1_Cr = [0.48, 40.53, 2.19, 0.55, 0.74, 0.66, 0.93, 0.37, .22]
x2_St = [12.57, 73.68, 11.13, 20.03, 20.29, 0.78, 4.64, 0.43, 1.08]

hair_data = np.array([x1_Cr, \
                      x2_St, \
                     ]).T

x1_Cr_mean = np.mean(hair_data[:, 0])
x2_St_mean = np.mean(hair_data[:, 1])

hair_data_cov = np.cov(hair_data.T)
hair_data_cov

# (a) 90% joint confidence ellipse
hair_data_mu = np.array([x1_Cr_mean, x2_St_mean])
hair_data_Sigma = np.array(np.array(hair_data_cov))

hair_data_mu_hat = np.average(hair_data, axis=0)

eigenvalues2, eigenvectors2 = np.linalg.eigh(hair_data_cov)
hair_data_S_square_root = eigenvectors2 @ np.diag(np.sqrt(eigenvalues2)) @ eigenvectors2.T

hair_data_df = pd.DataFrame(hair_data)

hair_data_radius_square = ((n1-1)*p1/((n1-p1)*n1))* f.ppf(0.90, p1, n1-p1)
hair_data_radius_square

hair_data_ellipsoid = [hair_data_S_square_root @ point + hair_data_mu_hat
                             for point in (np.sqrt(hair_data_radius_square) * circle)
                            ]

ax2 = hair_data_df.plot.scatter(0, 1, c="r", label="$\\vec x_i, i=1,...,n$")
pd.DataFrame(hair_data_ellipsoid).plot.scatter(0, 1, ax=ax2)
```
